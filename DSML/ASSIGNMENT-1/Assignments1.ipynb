```python
import numpy as np
import pandas as pd
```

```python
# a. read data from different formats(like csv, xls)
dataset=pd.read_csv("/content/Heart.csv")
dataset.head()
```

```python
# b. indexing and selecting data, sort data, describe attributes of data
AgeandThal=dataset[["Age","Thal"]]
AgeandThal.iloc[10:25,:]
```

```python
# Sort Data by Age
dataset.sort_values(["Age"],
                    axis=0,
                    ascending=[True],
                    inplace=True)
dataset.head()
```

```python
# describe attributes of data
dataset.describe()
```

```python
# checking data types of each column
dataset.info()
```

```python
# d. counting unique values of data
unique_value=dataset["Chol"].nunique()
print((unique_value))
```

```python
# e. format of each column,
dataset.applymap(type)
```

```python
# f. converting variable data type (e.g. from long to short, vice versa)
dataset['Oldpeak'] = dataset['Oldpeak'].astype(int)
dataset.applymap(type)
```

```python
# g. identifying missing values and fill in the missing values
missing_values = dataset.isnull().sum()
print(missing_values)
dataset.fillna(0, inplace=True)
```
